```typescript
import * as fs from 'node:fs/promises'

const file = await fs.readFile('input', 'utf8')

let id = 0
let image = []
let freeState:Boolean = false
let freeCount:Number = 0

// create array
for(const x of file) {
    const size = Number(x)
    
    if (!freeState) {
        for(let y = 0; y < size; ++y) {
            image.push(id)
        }
        ++id
        //if (id > 9) id = 0
        freeState = true
    } else {
        for (let z = 0; z < size; ++z) {
            image.push('.')
            ++freeCount;
        }
        freeState = false
    }
}

console.log(image.join('').length, freeCount)

// compact array, First try!
for (let a = 0; a < freeCount; ++a) {
    const lastPos = image.length - a - 1
    const last = image[lastPos]
    image[lastPos] = '.'

    for (let b = 0; b < image.length; ++b) {
        if (image[b] === '.') {
            image[b] = last
            break;
        }
    }
    //console.log(image.join(''))
}

// create checksum
let checksum = 0;
for(let pos =0; pos < image.length; ++pos) {
    const value = image[pos]
    if (value === '.') break

    const num = Number(value)
    checksum += pos * num
}

// Answers:
// 5630566118 -- too low
// 6332189866718 ✔️
console.log(checksum)
```

```typescript
import * as fs from 'node:fs/promises'

const file = await fs.readFile('sample', 'utf8')

let id = 0
let image = []
let freeState:Boolean = false
let freeCount:Number = 0

// create array
for(const x of file) {
    const size = Number(x)
    
    if (!freeState) {
        for(let y = 0; y < size; ++y) {
            image.push(id)
        }
        ++id
        freeState = true
    } else {
        for (let z = 0; z < size; ++z) {
            image.push('.')
            ++freeCount;
        }
        freeState = false
    }
}

console.log(image.join(''), id)
for (;freeCount > 0; --freeCount) {
    --id
    // get count of id
    let count = 0
    let oldPos = 0
    for (let x = 0; x < image.length; ++x) {
        if (image[x] === id) {
            ++count
            if (oldPos === 0) {
                oldPos = x;
            }
        }
    }
    
    // Find empty space
    let pos = 0
    let mvCnt = 0
    let found = false
    for (let y = 0; y < image.length; ++y) {
        if (image[y] === '.') {
            mvCnt++
        } else {
            if (mvCnt >= count && pos === 0) {
                pos = y - mvCnt
                found = true
            }
            mvCnt = 0
        }
    }

    if (!found) continue

    console.log(count, oldPos, pos, mvCnt)
    
    // move the items
    for (let z = 0; z < count; ++z) {
        image[z + pos] = image[z + oldPos]
        image[z + oldPos] = '.'
    }

    console.log(image.join(''), pos, count)
}

// create checksum
let checksum = 0;
for(let pos =0; pos < image.length; ++pos) {
    const value = image[pos]
    if (value === '.') break

    const num = Number(value)
    checksum += pos * num
}

// Answers:
// 5630566118 -- too low
// 6332189866718 ✔️
console.log(checksum)
```
