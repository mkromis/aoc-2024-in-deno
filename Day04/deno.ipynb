```typescript
import * as fs from 'node:fs/promises'

const file = (await fs.readFile('input', 'ascii')).toString()
let data:string[][] = [] // Save data from file
const word = 'XMAS'
let count = 0

// populate data
for(const idx of file.split('\n')) {
    // Create array then push to main
    let row = []
    for(const char of idx) {
        row.push(char)
    }
    data.push(row)
}

for(let x = 0; x < data.length; ++x) {
    for(let y = 0; y < data[x].length; ++y) {
        let found = false
        // Find the first X
        if (data[x][y] === word[0]) {
            // find forward
            if (y < data[0].length - 2 && (data[x][y] + data[x][y + 1] + data[x][y + 2] + data[x][y + 3]) === word) {
                console.log('F ' + x + ':' + y)
                count++
                found = true;
            }
            
            // Reverse
            if (y > 2 && (data[x][y] + data[x][y-1] + data[x][y-2] + data[x][y-3]) === word) {
                console.log('R ' + x + ':' + y)
                count++
                found = true;
            }

            // Down
            if (x < data[x].length - 3 && (data[x][y] + data[x+1][y] + data[x+2][y] + data[x+3][y]) === word) {
                console.log('D ' + x + ':' + y)
                count++
                found = true;
            }

            // Up
            if (x > 2 && (data[x][y] + data[x-1][y] + data[x-2][y] + data[x-3][y]) === word) {
                console.log('U ' + x + ':' + y)
                count++
                found = true;
            }

            if (x < data.length - 3 && y < data[0].length - 2 && (data[x][y] + data[x+1][y + 1] + data[x+2][y + 2] + data[x+3][y + 3]) === word) {
                console.log('DR ' + x + ':' + y)
                count++
                found = true;
            }

            if (x > 2 && y < data.length - 3 && (data[x][y] + data[x-1][y + 1] + data[x-2][y + 2] + data[x-3][y + 3]) === word) {
                console.log('UR ' + x + ':' + y)
                count++
                found = true;
            }

            if (x < data[x].length - 3 && y > 2 && (data[x][y] + data[x+1][y-1] + data[x+2][y - 2] + data[x+3][y - 3]) === word) {
                console.log('DL ' + x + ':' + y)
                count++
                found = true;
            }

            if (x > 2 &&  y > 2 && (data[x][y] + data[x-1][y-1] + data[x-2][y-2] + data[x-3][y-3]) === word) {
                console.log('UL ' + x + ':' + y)
                count++
                found = true;
            }
        }
    }
}

// sample: 18
// input: 2718
console.log(`Count: ${count}`)
```
