```typescript
import * as fs from 'node:fs/promises'

const file = await fs.readFile('input', 'utf8')
const lines = file.split('\n')

interface Steps {
    opcode: Number,
    operand: Number
}

let A:number = 0
let B:number = 2024
let C:number = 43690

let program:number[] = []
let steps:Steps[] = []


// Register A
A = Number(lines[0].split(':')[1]) 

// split program line
for(let x of lines[4].split(':')[1].split(',')) {
    program.push(Number(x))
}

// combine into steps
for (let x = 0; x < program.length - 1;) {
    let code = program[x++]
    let op = program[x++]
    steps.push({
        opcode: code,
        operand: op
    })
}

console.log(steps.length)

let output = ''
for(let x = 0; x < steps.length; ++x) {
    // use this if jumping
    const step = steps[x] 
    let combo = 0

    if (step.operand < 4) combo = step.operand
    else if (step.operand === 4) combo = A
    else if (step.operand === 5) combo = B
    else if (step.operand === 6) combo = C

    // Do the operation
    switch (step.opcode) {
        case 0:  // adv -- performs division
            combo = Math.pow(2,combo)
            A = parseInt(A /combo)
            break;
        case 1: // bxl -- bitwise
            B = B ^ step.operand
            break;  
        case 2: // bst -- combo
            B = combo % 8
            break;
        case 3:
            if (A !== 0) x = step.operand - 1
            break;
        case 4:
            B = B ^ C
            break;
        case 5:
            output += ',' + (combo % 8)
            break;
        case 6:
            combo = Math.pow(2,combo)
            B = parseInt(A /combo)
            break;
        case 7:
            combo = Math.pow(2,combo)
            C = parseInt(A /combo)
            break;
        default:
            throw Error('Invalid Opcode', step.opcode)
    }
}
//console.log('Register', B)
// Solution: 7,0,7,3,4,1,3,0,1
console.log('Output:', output.substring(1))
```
