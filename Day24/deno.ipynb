```typescript
const file = await Deno.readTextFile('input')
const lines = file.split('\n')

const state = new Map<string, boolean>()
const lookup = new Map<string, string>()
let isOpcode = false

function getValue(item:string) : boolean {
    // Cache
    if (state.has(item)) {
        return state.get(item)
    }

    // first get the record
    const value = lookup.get(item)
    const split = value?.split(' ')
    const first = getValue(split[0]) 
    const op = split[1] // Op
    const second = getValue(split[2])

    let result = false
    switch (op) {
        case 'OR':
            result = first | second
            break;
        case 'XOR':
            result = first ^ second
            break;
        case 'AND':
            result = first & second
            break;
                
        default:
            throw new Error()
            break;
    }
    state.set(item, result)
    return state.get(item)
}

for(const line of lines) {
    if (line === '') {
        isOpcode = true
        continue
    }
    
    if (!isOpcode) {
        // Store the state
        let split = line.split(':')
        state.set(split[0], split[1].trim() === '1' ? true : false)
    } else {
        let split = line.split('->')
        lookup.set(split[1].trim(), split[0].trim())
    }
}

let keys = []
// Get list of items to find
for (const item of lookup.keys()) {
    if (item.startsWith('z')) {
        keys.push(item)
    }
}
keys = keys.sort();
console.log(keys)

// Get results
for(const key of keys) {
    console.log(key, getValue(key))
}

// Answer 55730288838374
```
