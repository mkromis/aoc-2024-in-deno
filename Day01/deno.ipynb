```typescript
// Part 2
import * as fs from 'node:fs/promises'
import * as path from 'node:path'

const filename =  path.resolve('input')
const input = (await fs.readFile(filename)).toString()
const lines = input.split('\n')

// setup for number compare
let left:Number = []
let right:Number = []

// split values
for(const index in lines) { // index    
    const line = lines[index] // get line
    const split = line.split(" ") 

    // convert to number
    left.push(Number(split[0]))
    right.push(Number(split.at(-1)))
}

// sort items
left = left.sort((a: Number,b: Number) => a - b)
right = right.sort((a: Number,b: Number) => a - b)

let diff = 0
let lookup = new Map<Number, Number>()

// Calulate how many duplicate keys
for(const index in lines) { // index    
    const rightNum = Number(right[index])
    let count = 0;
    if (lookup.has(rightNum)) {
        count = lookup.get(rightNum);
    }

    ++count
    lookup.set(rightNum, count)
}

// calulate the similarity
let result = 0
for(const index in lines) {
    const leftNum = left[index];
    if (lookup.has(leftNum)) {
        result += leftNum * lookup.get(leftNum)
    }
}
console.log("Uniques:", lookup.size)
console.log(result)  // Answer

// Sample is 31
// input is 24349736
```
